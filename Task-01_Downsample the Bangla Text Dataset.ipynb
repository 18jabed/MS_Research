```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
text_data_path = 'F:/MS Thesis/MultiModel/Text.csv'
df = pd.read_csv(text_data_path)

# Display basic info about the dataset
print("Dataset Info:")
print(df.info())
print("\nFirst 5 rows:")
print(df.head())

# Plot original class distribution
if 'Emotion' in df.columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(x='Emotion', data=df, palette='viridis')
    plt.title('Original Class Distribution of Emotions')
    plt.xlabel('Emotion Classes')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
else:
    print("Column 'Emotion' not found in the dataset. Please check the column names.")

# Downsample the text data to 6000 samples (ensure the same size as voice dataset)
df = df.sample(n=6000, random_state=42)

# Plot downsampled class distribution
if 'Emotion' in df.columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(x='Emotion', data=df, palette='plasma')
    plt.title('Downsampled Class Distribution of Emotions')
    plt.xlabel('Emotion Classes')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
else:
    print("Column 'Emotion' not found in the dataset. Please check the column names.")

# Save the downsampled file
save_path = 'F:/MS Thesis/Research/Text.csv'
df.to_csv(save_path, index=False)

# Display value counts in table format
if 'Emotion' in df.columns:
    class_counts = df['Emotion'].value_counts()
    print("\nClass Distribution Table (Downsampled):")
    print(class_counts)
```
